```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("data101.csv")
data
```

```python
data = data[['Country', 'Year', 'overallreadiness', 'overallvulnerability']]
data
```

```python
data['readiness_pct_change'] = data['overallreadiness'].pct_change()
data['vulnerability_pct_change'] = data['overallvulnerability'].pct_change() 
data
```

```python
data.info() 
```

```python
data['Year'] = pd.to_datetime(data['Year'], format="%Y")
data
```

```python
sliced_data = data[['Country', 
                    'Year', 
                    'readiness_pct_change', 
                    'vulnerability_pct_change', 
                    'overallreadiness', 
                    'overallvulnerability']]
```

```python
def extract_country(country): 
    df = sliced_data[sliced_data['Country'] == country]
    return df
```

```python
# creating functions for plotting the 
def plot_country_readiness(country, color): 
    df = extract_country(country)
    plt.plot(df['Year'], df['readiness_pct_change'], color=color, label=country)

def plot_country_vulnerability(country, color): 
    df = extract_country(country)
    plt.plot(df['Year'], df['vulnerability_pct_change'], color=color, label=country)
```

```python
countries = ['Brunei Darussalam',
 'Cambodia',
 'Indonesia',
 'Lao PDR',
 'Malaysia',
 'Myanmar',
 'Philippines',
 'Singapore',
 'Thailand',
 'Viet Nam'] 
```

```python
# time series graph showing the rate of change in the readiness index 

#plotting the axes and the titles 
plt.title('Rate of Change in Readiness Index')
plt.xlabel('Year')
plt.ylabel('Rate of Change')

# plotting each individual line w/ premade functions above 
plot_country_readiness('Cambodia', 'green')
plot_country_readiness('Indonesia', 'red')
plot_country_readiness('Lao PDR', 'blue')
plot_country_readiness('Malaysia', 'cyan')
plot_country_readiness('Philippines', 'darkslategray')
plot_country_readiness('Singapore', 'm')
plot_country_readiness('Thailand', 'yellow')
plot_country_readiness('Viet Nam', 'orange')
plt.legend()


# saving the image 
plt.savefig("readiness.png")
plt.show() 
```

```python
#plotting the axes and the titles 
plt.title('Rate of Change in Overall Vulnerability Index')
plt.xlabel('Year')
plt.ylabel('Rate of Change')

# plotting the lines 
plot_country_vulnerability('Cambodia', 'green')
plot_country_vulnerability('Indonesia', 'red')
plot_country_vulnerability('Lao PDR', 'blue')
plot_country_vulnerability('Malaysia', 'cyan')
plot_country_vulnerability('Philippines', 'darkslategray')
plot_country_vulnerability('Singapore', 'm')
plot_country_vulnerability('Thailand', 'yellow')
plot_country_vulnerability('Viet Nam', 'orange')


plt.legend()

plt.savefig('vulnerability.png')

plt.show()

```
